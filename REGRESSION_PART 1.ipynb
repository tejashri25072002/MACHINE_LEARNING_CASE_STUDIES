Regression

It is method for understanding a relationship between dependent and independent variables.

It is used on continuous dataset

**2 Types of regression:**
1) Simple Regression
------------Linear simple regression
------------Non-linear simple regression

--If one feature is mapped for target then it is **simple
  regression**.
--If these selected feature nad target have linear
  relationship then it is called as **simple linear regression**.
--If it is non-linearity between features and target then it
  is called as **simple non-linear regresssion**.

2) Multiple Regression
------------Linear multiple regression
------------Non-linear multiple regression

--If more than one feature are mapped with target then it
  will be **multiple regression**.
--If these selected feature nad target have linear
  relationship then it is called as **multiple linear regression**
--If it is non-linearity between features and target then it
  is called as **multiple non-linear regression**.

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
x=np.array([1,2,3,4])
slope=0.5
intercept=4

y = slope*x+intercept    #Formula

y
```

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
area=np.array([400,800,1000,1200,1500,2000])    # actual x
price=np.array([4,6,9,12,16,21])
plt.scatter(area,price)
```

```python
ypred=area*slope+intercept
```

```python
area_m=np.mean(area)
price_m=np.mean(price)
slope=np.sum((area-area_m)*(price-price_m))/np.sum((area-area_m)**2)
intercept=price_m-(slope*area_m)
```

```python
ypred=area*slope+intercept
ypred
```

```python
plt.figure()
plt.scatter(area,price)
plt.plot(area,ypred)
plt.show()

# We can observe that there is linear relationship between feature and target
# But the correlation value will not be close 1
```
