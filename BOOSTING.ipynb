```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('bank.csv')
data.head()
```

```python
# Define training and testing data

X=data.drop('deposit',axis=1)
y=data['deposit']
```

```python
# train test split data

xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
# Fitting of weak classifier

# weak learner or week classifier mean the algorithm that consist of error or mistake

dt_clf=DecisionTreeClassifier(max_depth=1,random_state=0)
dt_clf.fit(xtrain,ytrain)
dt_score=dt_clf.score(xtest,ytest)
print("decision tree score:",dt_score)
```

```python
#fitting of weakclassifire with Adaboost

from sklearn.ensemble import AdaBoostClassifier
ada_clf=AdaBoostClassifier(base_estimator=dt_clf,random_state=0)

#n_estimator=it controls the number of weak learners
#base estimator--it helps to specify different ML algorithm

ada_clf.fit(xtrain,ytrain)
ada_score=ada_clf.score(xtest,ytest)
print("ada boost score is ",ada_score)
```
