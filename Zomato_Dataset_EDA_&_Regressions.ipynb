```python
# Remove junk character

df['Latitude'].replace('.',np.nan,inplace=True)
```

```python
df['Latitude'].unique()
```

```python
df['Currency'].replace('?',np.nan,inplace=True)
```

```python
df['Currency'].unique()
```

```python
# Check col datatype

df['Latitude'].dtype
```

```python
# convert into float

df['Latitude']=df['Latitude'].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Latitude'].fillna(df['Latitude'].mean(),inplace = True)
```

```python
# separating feature and target

x = df.iloc[:,:-1]     # all col except price
y = df.iloc[:,-1]      # all rows of last col
```

```python
y.head()
```

```python
# Find out outliers

sns.boxplot(data=x,x=y)
plt.grid()
plt.show()
```

```python
sns.boxplot(data=x,x=y,y='Currency')
plt.grid()
plt.show()
```
