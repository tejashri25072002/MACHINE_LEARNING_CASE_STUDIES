```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
area=np.array([600,800,1000,1200,1500,2000])
price=np.array([4,6,9,12,16,21])#lakhs
print(area)
print(price)
```

```python
#visualizing above data using scatter plot
plt.scatter(area,price,c='green',marker='*',s=50)
plt.title("area vs price")
plt.xlabel('area in sqft',c='blue')
plt.ylabel('price in lakhs',c='pink')
plt.show()
```

```python
#line plot
plt.plot(area,price,c='red')
#plt.scatter(area,price,c='black',marker='*')
plt.title("area vs price")
plt.xlabel('area in sqft',c='blue')
plt.ylabel('price in lakhs',c='pink')
plt.show()
```

```python

#two in one graph
#line plot
plt.plot(area,price,c='red')---graph1
plt.scatter(area,price,c='black',marker='*')---graph2
plt.title("area vs price")
plt.xlabel('area in sqft',c='blue')
plt.ylabel('price in lakhs',c='pink')
plt.show()
```
