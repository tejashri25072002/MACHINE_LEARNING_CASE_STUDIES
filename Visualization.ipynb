```python
# 1. Scatter Plot

plt.figure(figsize=(5,5))
plt.title("Width v/s Price of cars")
x = car['width']
y = car['price']
plt.scatter(x,y)
plt.xlabel("Width of Car")
plt.ylabel("Price of Car")
plt.show()
```

```python
# 2.Histogram

plt.hist(car['body-style'])
plt.xlabel('Body Style')
plt.ylabel('No. of Cars')
plt.title("Body style v/s No. of cars")
c = car['body-style'].value_counts()

plt.show()
```

```python
# 3. Bar Plot

car['fuel-type'].value_counts().plot(kind='bar')
plt.xlabel('Fuel Type')
plt.ylabel('No. of Cars')
plt.title("Fuel type v/s No.of cars")
c = car['fuel-type'].value_counts()
plt.show()
```

```python
# 4. Pie plot

#explode ---> indicate space from center

car['body-style'].value_counts().plot(kind='pie',autopct='%.2f%%',explode=(0,0.3,0,0.5,0))
plt.show()
```

```python
# 5. Count Plot

plt.figure(figsize=(5,5))
sns.countplot(data=car,x='body-style')
f=car['body-style'].value_counts()
plt.yticks(f)
plt.show()
```

```python
# 6. Heatmap  --> useful for check correlation

sns.heatmap(car.corr(),annot=True)
plt.show()
```

```python
# 7. Boxplot

X = car.iloc[:,:-1]  # retrive all col but except price
#X = car['symboling','city-mpg'.......]     ---> all col names
target = car.iloc[:,-1]   # Price col display
# target= car['price']
sns.boxplot(data=X,x=target,y='make')
plt.grid()
```
