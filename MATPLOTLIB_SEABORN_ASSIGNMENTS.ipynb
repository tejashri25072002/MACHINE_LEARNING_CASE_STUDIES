```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

x = np.arange(0, 100)
y = x * 2
z = x**2

ax1.plot(x, y, label='y = 2x')
ax2.plot(x, y, label='y = 2x')

ax1.set(xlabel='X-axis', ylabel='Y-axis', title='Plot on ax1')
ax2.set(xlabel='X-axis', ylabel='Y-axis', title='Plot on ax2')

plt.tight_layout()

plt.show()
```

***ASSIGNMENT 3***

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

x = np.arange(0, 100)
y = x * 2
z = x**2

ax1.plot(x, y, label='y = 2x')
ax1.set(xlabel='X-axis', ylabel='Y-axis', title='Main plot')

ax2.plot(x, z, label='z = x^2')
ax2.set_xlim(20, 60)  # Set x-axis limits for zoomed view
ax2.set_ylim(0, 4000)  # Set y-axis limits for zoomed view
ax2.set(xlabel='X-axis', ylabel='Y-axis', title='Zoomed inset')

plt.tight_layout()

plt.show()
```
