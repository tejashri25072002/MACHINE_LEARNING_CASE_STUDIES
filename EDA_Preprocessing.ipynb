***#Skewness***

iT is measure of asymetry
It tells us how datapoints are distributed 3 types
1) positive
2) Negative
3) Zero

We reduce skewness by using 2 methods
1) Longvalue
2) Squareroot value

```python
from scipy.stats import skew
```

```python
colname = x.select_dtypes(['int64','float64']).columns
colname
```

```python
#apply skew function on one column

skew(x['normalized-losses'])
```

```python
# apply for loop for calculating skew for all col

for col in x[colname]:
  print(col)
  print(skew(x[col]))
  plt.figure()
  sns.distplot(x[col])
  plt.show()
```

Before removing skewness we need to check correlation of features with target

We need to follow ***2 rules***:
1) If the column is skewed  but showing good correlation with
   target then ***do not remove skewness***
2) if the col which is skewed has some negative values dont  
   not skewness even showing poor correlation with target

```python
# removing skewness by using logvalue method

x['normalized-losses']= np.log(x['normalized-losses'])
skew(x['normalized-losses'])
```
