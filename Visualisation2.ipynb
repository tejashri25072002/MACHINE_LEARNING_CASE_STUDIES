```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
area=np.array([600,800,1000,1200,1500,2000])
price=np.array([4,6,9,12,16,21])#lakhs
print(area)
print(price)
```

```python
#visualizing above data using scatter plot
plt.scatter(area,price,c='green',marker='*',s=50)
plt.title("area vs price")
plt.xlabel('area in sqft',c='blue')
plt.ylabel('price in lakhs',c='pink')
plt.show()
```

```python
#line plot
plt.plot(area,price,c='red')
#plt.scatter(area,price,c='black',marker='*')
plt.title("area vs price")
plt.xlabel('area in sqft',c='blue')
plt.ylabel('price in lakhs',c='pink')
plt.show()
```

```python
#two in one graph
#line plot
plt.plot(area,price,c='red')
plt.scatter(area,price,c='black',marker='*')
plt.title("area vs price")
plt.xlabel('area in sqft',c='blue')
plt.ylabel('price in lakhs',c='pink')
plt.show()
```

```python
#scatter plot using seaborn---------it works on dataframe
data={
    'area': [600,800,1000,1200,1500,2000],
    'price': [4,6,9,12,16,21]#lakhs
}
```

```python
df=pd.DataFrame(data)
df
```

```python
X=df['area']
Y=df['price']
sns.scatterplot(data=df,x=X,y=Y)
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic['who'].value_counts()
```

```python
titanic.info()
```

```python
#count plot on class col
sns.countplot(data=titanic,x='class')
f=titanic['class'].value_counts()
plt.yticks(f)#-----indicate exact count of people in each class
plt.show()
```

```python
#count plot on'who' column
sns.countplot(data=titanic,x='class',hue='who')
plt.show()
```

```python
#countplot
sns.countplot(data=titanic,x='survived',hue='pclass')
```

```python
#pie chart
titanic['pclass'].value_counts().plot(kind='pie',autopct='%.2f%%')
plt.show()
```
