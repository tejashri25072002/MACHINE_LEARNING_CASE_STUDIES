```python
'''
75 % and above = A grade
74 - 60        = B grade
59 - 35        = c grade
Below 35       = Fail
p = marks/total*100
'''

#Function to calculate the grade

def grade(marks):
  p = (marks/200)*100
  if(p > 75):
    return 'A'
  elif(p >= 60 and p < 74):
    return 'B'
  elif(p >= 35 and p < 60):
    return 'c'
  else:
    return 'Fail'
```

```python
df['Grade'] = df['Total'].apply(grade)
df
```
